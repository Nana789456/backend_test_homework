```python
def fun(*args): # args - это параметр, который содержит в себе ВСЕ переданные НЕ именнованные (позиционные) аргументы в виде кортежа
    print(type(args))
    print(args)
    print(args[1])
    return
```

```python
fun(1, 2, 3)
```

```python
def temer(fun):
    def wrapper(*args):
        print("Програма начала выполнение")
        fun(*args)
        print("Програма закончила выполнение за ...")
    return wrapper
```

```python
@temer
def fun1(name, age, sex): 
    pass

@temer
def fun2(name, age): 
    pass

# wrapper = temer(fun1)
# wrapper()
```

```python
"""
args - это параметр, который содержит в себе ВСЕ переданные НЕ именнованные (позиционные) аргументы в виде кортежа
kwargs - это параметр, который содержит ВСЕ именованные аргументы
"""
def fun(*args, **kwargs): 
    print(type(args))
    print(args)

    print(type(kwargs))
    print(kwargs)
    return
```

```python
fun(1, 2, 3)
```

```python
fun(1, 2, 3, name="Bob", age=26)
```

```python
fun(1, 2,  name="Bob", 3, age=26)
```

```python
# fun(1, 2,  name="Bob", 3, age=26)
def fun(*args, **kwargs, *args, **kwargs): 
    print(type(args))
    print(args)

    print(type(kwargs))
    print(kwargs)
    return


fun(1, 2,  name="Bob", 3, age=26)

"""
Нам НЕЛЬЗЯ смешивать позиционные и именованные аргументы, т.к. если их пермешать, то Python НЕ поймёт что из них относится к args, а что к kwargs параметру
"""
```

```python
def fun(*args, **kwargs): 
    print(type(args))
    print(args)

    print(type(kwargs))
    print(kwargs)
    return


fun(1, 2,  3, name="Bob", age=26)
```
