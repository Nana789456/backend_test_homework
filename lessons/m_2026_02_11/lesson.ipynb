```python
from dataclasses import dataclass

@dataclass
class User:
    name: str
    age: int
    email: str = "no@email.com"  # значение по умолчанию

# Создаём объект
user = User(name="Иван", age=25)
print(user)  # User(name='Иван', age=25, email='no@email.com')
```

```python
!pip install pydantic

from pydantic import BaseModel, EmailStr, Field

class UserPydantic(BaseModel):
    name: str = Field(min_length=2, max_length=50)
    age: int = Field(ge=0, le=150)  # greater or equal, less or equal
    email: EmailStr | None = None  # автоматическая валидация email

# Всё ок
user1 = UserPydantic(name="Анна", age=30, email="anna@mail.ru")

# Ошибка валидации!
try:
    user2 = UserPydantic(name="А", age=200, email="not-email")
except Exception as e:
    print(e)  # Подробно расскажет, что не так

print(user1.model_dump())  # Сериализация в dict
print(user1.model_dump_json())  # Сериализация в JSON
```
