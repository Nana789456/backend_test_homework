```python
# Из чего состоит HTTP запрос ?

"""
- Адрес запроса  http://localhost:8000/items/1/?age=10

- Тип запроса (GET/POST/PUT/DELETE запросы)
(!) Мы сообщаем серверу тип нашего дейтсвия, чтобы было понятно что мы хотим в принципе сделать

- path параметры (параметры пути), например 1
(!) Параметры пути конкретизируют положение элементов нашей доменной области (нашей бизнес логики)
Например в https://github.com/settings/security-log:
-- https://github.com -> корень сайта
-- settings, security-log -> параметры пути

- query параметры (параметры запроса)
(!) Параметры запроса конкретизируют (дают контекст) того что нужно сделать по адресу запроса
с конкретными элементами нашей доменной области

- Тело запроса (request body) 
(!) часть HTTP-запроса, содержащая передаваемые на сервер данные 
(JSON, XML, HTML-формы), следующая после заголовков. 
(!) Оно используется в методах POST, PUT, PATCH для отправки, создания или обновления информации (наших объектов). 
(!) Тело обычно отсутствует в GET и DELETE.

- Заголовки (headers)
(!) Заголовки используются для передачи дополнительной информации, 
такой как тип контента (Content-Type), параметры аутентификации (Authorization), 
кодировки и кеширования (Cache-Control, инфорамация об одресате (точная)
(!) В отличае от информации в параметрах и теле запроса в них хранится дополнительной информация 
для корректного выполнения запроса.
Например чтоббы только авторизованный пользователь мог сделать запрос, тип парсинга запроса.

- Ответ сервера
(!) Кот ответа
(!) Информация, которую нам возвращает сервер на наш запрос (может быть в виде JSON)


https://practicum.yandex.ru/blog/chto-takoe-rest-api-i-kak-rabotaet/
"""
```
