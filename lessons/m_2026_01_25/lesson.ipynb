```python
users_marks = {
    'user1': [3, 1, 2],
    'user2': [5, 5, 5],
    'user3': [5, 4, 5],
    'user4': [2, 1, 2],
    'user5': [3, 2, 2]
}
```

```python
# Шаг 1: Исходный вариант

def get_average_value(key, users_marks):
    marks = users_marks[key]
    sum_marks = sum(marks)
    number_marks = len(marks)
    average = sum_marks / number_marks
    return average

def get_min_average_value(users_marks):
    return min(users_marks, key=get_average_value)

get_min_average_value(users_marks)
```

```python
# Шаг 2: Добавляем параметр
def get_average_value(key, users_marks):  # Теперь принимает словарь
    marks = users_marks[key]
    sum_marks = sum(marks)
    number_marks = len(marks)
    average = sum_marks / number_marks
    return average

def get_min_average_value(users_marks):
    # Нужно адаптировать вызов под новую сигнатуру
    return min(users_marks, 
               key=lambda k: get_average_value(k, users_marks))

get_min_average_value(users_marks)
```

```python
# Шаг 3: Убираем промежуточные переменные
def get_average_value(key, users_marks):
    marks = users_marks[key]
    return sum(marks) / len(marks)  # Сразу возвращаем результат

def get_min_average_value(users_marks):
    return min(users_marks,
               key=lambda k: get_average_value(k, users_marks))

get_min_average_value(users_marks)
```

```python
# Шаг 4: Встраиваем логику прямо в lambda
def get_min_average_value(users_marks):
    return min(users_marks,
               key=lambda k: sum(users_marks[k]) / len(users_marks[k]))

get_min_average_value(users_marks)
```
